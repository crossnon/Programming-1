```python
# Q1
F = float(input("Enter the temperature in Fahrenheit: "))

if F <= 32:
    print("The result is 'cold'.")
else:
    C = (F - 32) * 5 / 9
    print("The temperature in Celsius is:", C)
```

```python
# Q2
start = int(input("Enter the starting number of the range: "))
end = int(input("Enter the ending number of the range: "))
sum_of_squares = 0

for number in range(start, end + 1):
    if number % 2 == 0:
        sum_of_squares += number ** 2

print("Sum of the squares of even numbers within the range", start, "to", end, "is:", sum_of_squares)
```

```python
# Q3
# Take input for four numbers
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))
num3 = int(input("Enter the third number: "))
num4 = int(input("Enter the fourth number: "))

# Initialize variables for biggest and smallest numbers
biggest = smallest = num1

# Find the biggest and smallest numbers
if num2 > biggest:
    biggest = num2
elif num2 < smallest:
    smallest = num2

if num3 > biggest:
    biggest = num3
elif num3 < smallest:
    smallest = num3

if num4 > biggest:
    biggest = num4
elif num4 < smallest:
    smallest = num4

# Print the biggest and smallest numbers
print("The biggest number is:", biggest)
print("The smallest number is:", smallest)
```

```python
# Q4
# Input time of operation for each element
burner_time = float(input("Enter the time of operation for the burner [hours]: "))
computer_time = float(input("Enter the time of operation for the computer [hours]: "))
ac_time = float(input("Enter the time of operation for the air conditioner [hours]: "))
bulb_time = float(input("Enter the time of operation for the bulbs [hours]: "))

# Calculate energy consumption in kWh for each element
burner_energy = 1.2 * burner_time
computer_energy = 0.6 * computer_time
ac_energy = 5 * ac_time
bulb_energy = 0.06 * bulb_time * 10  # 10 bulbs, each consuming 60 W

# Calculate total energy consumption in kWh
total_energy = burner_energy + computer_energy + ac_energy + bulb_energy

# Calculate cost of the electric bill
total_energy = total_energy * 30
if total_energy <= 100:
    cost = 0
elif total_energy <= 200:
    cost = (total_energy - 100) * 5
else:
    cost = 100 * 5 + (total_energy - 200) * 10



# Print the cost of the electric bill for the month
print("The cost of the electric bill for the month is: $", cost)
```

```python
# Q5
num = int(input("Enter a number: "))

# Initialize variables
sum_of_digits = 0
temp = num

# Calculate sum of digits
while temp > 0:
    digit = temp % 10
    sum_of_digits += digit
    temp //= 10

print("Sum of digits of", num, "is:", sum_of_digits)
```

```python
# Q6
start = int(input("Enter the starting number: "))
end = int(input("Enter the ending number: "))

prime_numbers = []

for num in range(start, end+1):
    if num > 1:
        for i in range(2, num):
            if (num % i) == 0:
                break
        else:
            prime_numbers.append(num)

print("Prime numbers between", start, "and", end, "are:")
for prime in prime_numbers:
    print(prime, end=" ")
```
